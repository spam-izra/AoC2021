# Part 1



Теперь у нас дым растекается по пещере. Имеется карта высот дна пещеры (пример)
```
2199943210
3987894921
9856789892
8767896789
9899965678
```
где 9 - самая большая высота, 0 - минимальная высота.

Нужно найти все точки-минимумы (в примере 4). Точка-минимум - это позиция, которая окружена высотами большими, чем высота в текущей позиции. Вывести сумму оценок риска точек-минимума. Оценка риска - это значение высоты точки + 1.

Вывод: 15

````python
hmap = []

with open('test.txt') as f:
    for line in f:
        line = line.strip()
        if line == '': continue
        
        hmap.append(
            list(map(int, line)),
        )


directions = [
    ( 0, 1),
    ( 0,-1),
    ( 1, 0),
    (-1, 0),
]

N, M = len(hmap[0]), len(hmap)

risk = 0
for y, row in enumerate(hmap):
    for x, v in enumerate(row):
        
        for dx, dy in directions:
            if not (0 <= x + dx < N): continue
            if not (0 <= y + dy < M): continue
            if hmap[y + dy][x + dx] <= hmap[y][x]:
                break
        else:
            risk += v + 1
            print("Minimum", (x, y))
print(risk)
with open("output1.txt", "w") as f:
    print(risk, file=f)
````

# Part 2

Теперь нужно найти самый большой бассейн (область из которой все будет скатываться в точку-минимум). Точки с высотой 9 не являются частью бассейнов, остальные точки являются частью только одного бассейна. Размер бассейна - это количество точек в нем. В примере 4 бассейна. Вывод - это произведение размеров трех наибольших бассейнов.

Вывыод: 1134

````python
hmap = []

with open('input.txt') as f:
    for line in f:
        line = line.strip()
        if line == '': continue
        
        hmap.append(
            list(map(int, line)),
        )


directions = [
    ( 0, 1),
    ( 0,-1),
    ( 1, 0),
    (-1, 0),
]

N, M = len(hmap[0]), len(hmap)

mins = []
for y, row in enumerate(hmap):
    for x, v in enumerate(row):
        
        for dx, dy in directions:
            if not (0 <= x + dx < N): continue
            if not (0 <= y + dy < M): continue
            if hmap[y + dy][x + dx] <= hmap[y][x]:
                break
        else:

            mins.append((x, y))

basins = [0 for _ in mins]
print(mins)


bmap = [
    [None for _ in range(N)]
    for _ in range(M)
]

for i, (x, y) in enumerate(mins):
    queue = [(x, y)]
    bmap[y][x] = i
    while queue:
        x, y = queue.pop(0) 
        basins[i] += 1
        
        for dx, dy in directions:
            if not (0 <= x + dx < N): continue
            if not (0 <= y + dy < M): continue
            if bmap[y + dy][x + dx] is not None:
                continue
            if hmap[y + dy][x + dx] == 9:
                continue
            bmap[y+dy][x+dx] = i
            queue.append((x+dx, y+dy))
            
print(basins)

for row in bmap:
    for c in row:
        print(c if c is not None else ' ', end='')
    print()
    

basins.sort()
answer = basins[-1] * basins[-2] * basins[-3]
print(answer)

with open("output2.txt", "w") as f:
    print(answer, file=f)
````
