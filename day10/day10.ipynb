# Part 1



Программа навигации была повреждена. 

```
[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]
```

Каждая строчка должна состоять из правильно открывающихся и закрывающихся скобок. Некоторые из строчек не повреждены - просто не закончены. Игнорируем пока такие строчки. 

Необходимо остановится на первом незакрытом символе. Каждый символ имеет свое количество очков
- ): 3 очков.
- ]: 57 очков.
- }: 1197 очков.
- \>: 25137 очков.

Вывод: 26397

````python
with open("test.txt") as f:
    lines = [line.strip() for line in f]
    
left = set("(<{[")
right = set(")>}]")
pairs = {
    ")": "(",
    "]": "[",
    ">": "<",
    "}": "{",
}
score = {
    ")": 3,
    "]": 57,
    "}": 1197,
    ">": 25137,
}
    
answer = 0
for line in lines:
    stack = []    
    for i, c in enumerate(line):
        if c in left:
            stack.append(c)
        elif c in right:
            if stack[-1] == pairs[c]:
                stack.pop()
            else:
                print(line)
                print(" "*i, "^", sep="")
                print("ERR:", c)
                answer += score[c]
                break
        
print(answer)

with open("output1.txt", "w") as f:
    print(answer, file=f)
````

# Part 2

Теперь мы можем отсеять все поврежденные строки и сосредоточится на неполных. Нужно корректно закрыть эти строки и посчитать очки зыкрывания.
- ): 1 очко.
- ]: 2 очка.
- }: 3 очка.
- \>: 4 очка.


При подсчете очков, при переходе к следуюшему символу, текущая сумма должна умножаться на 5. Например, строка `])}>` даст 294 очка
1. Начинаем с 0
2. 5 умножить на 0 плюс ](2) дает 2
3. 5 умножить на 2 плюс )(1) дает 11
4. 5 умножить на 11 плюс }(3) дает 58
5. 5 умножить на 58 плюс \>(4) дает 294

```
}}]])})] - 288957 total points.
)}>]}) - 5566 total points.
}}>}>)))) - 1480781 total points.
]]}}]}]}> - 995444 total points.
])}> - 294 total points.
```

После того, как найдены очки для каждой строки. Строку нужно отсортировать и взять центральное значение. Это и будет ответ

Вывод: 288957

````python
with open("input.txt") as f:
    lines = [line.strip() for line in f]
    
left = set("(<{[")
right = set(")>}]")
pairs = {
    ")": "(",
    "]": "[",
    ">": "<",
    "}": "{",
}
rpairs = {
    v: k for k, v in pairs.items()
}
score = {
    ")": 1,
    "]": 2,
    "}": 3,
    ">": 4,
}
    
answer = []
for line in lines:
    stack = []    
    for i, c in enumerate(line):
        if c in left:
            stack.append(c)
        elif c in right:
            if stack[-1] == pairs[c]:
                stack.pop()
            else:
                stack = []
                break
    if not stack:
        continue
        
    print(stack)
    end = "".join([rpairs[c] for c in reversed(stack)])
    
    tmp = 0
    for c in end:
        tmp = tmp * 5 + score[c]
    print(end)
    print(tmp)
    print()
    answer.append(tmp)
    
answer.sort()
print(answer)
answer = answer[len(answer) // 2]
print(answer)

with open("output2.txt", "w") as f:
    print(answer, file=f)
````
