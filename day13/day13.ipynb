# Part 1



Нам нужна термальная камера, но для работы с ней нужно ввести лицензионный ключ. Он зашифрован. Чтобы его расшифровать, нужно сложить листок бумаги согласно инструкции в документации. Сгиб идет либо влево, либо вверх. Нужно найти сколько точек будет видно после первого складывания.

Пример входных данных

```
6,10
0,14
9,10
0,3
10,4
4,11
6,0
6,12
4,1
0,13
10,12
3,4
3,0
8,4
1,10
2,14
8,10
9,0

fold along y=7
fold along x=5
```

Вывод: 17

````python
class Page:
    def __init__(self, dots):
        self.__dots = set(dots)
        
    @property
    def dots(self): return self.__dots.copy()
        
    def fold(self, line, coord):
        if line == 'x':
            tmp = set()
            for x, y in self.__dots:
                if x < coord:
                    tmp.add((x, y))
                elif x > coord:
                    tmp.add((2*coord - x, y))
            self.__dots = tmp
        elif line == 'y':
            tmp = set()
            for x, y in self.__dots:
                if y < coord:
                    tmp.add((x, y))
                elif y > coord:
                    tmp.add((x, 2*coord - y))
            self.__dots = tmp
        else:
            raise Exception(line)
        
        
    def __repr__(self):
        xMax = max(self.__dots, key=lambda x: x[0])[0]
        xMin = min(self.__dots, key=lambda x: x[0])[0]
        yMax = max(self.__dots, key=lambda x: x[1])[1]
        yMin = min(self.__dots, key=lambda x: x[1])[1]
        
        grid = [
            ['.' for x in range(xMin, xMax+1)]
            for y in range(yMin, yMax+1)
        ]
        
        for x, y in self.__dots:
            grid[y][x] = '#'
            
        r = ""
        for row in grid:
            r += "".join(row) + "\n"
        r += "\n"
        return r
        
        
with open("input.txt") as f:
    step = 'dots'
    dots = []
    cmds = []
    for line in f:
        line = line.strip()
        if line == "":
            step = 'cmd'
            continue
        if step == 'dots':
            x, y = line.split(',')
            dots.append((int(x), int(y)))
        elif step == 'cmd':
            cmd, x = line[11:].split("=")
            cmds.append((cmd, int(x)))

p = Page(dots)
p.fold(*cmds[0])
answer = len(p.dots)

with open('output1.txt', 'w') as f:
    print(answer)
    print(answer, file=f)
````
